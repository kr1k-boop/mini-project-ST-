```python
import numpy as np
import pandas as pd

def distance(a,b):
       
    #distance=1-np.dot(a,b)/np.sqrt(np.dot(a,a)*np.dot(b,b))
    return abs(a-b)

def Similarity(Id1, Id2):
    a = df.iloc[Id1]
    b = df.iloc[Id2]
    
    if abs(a['year']-b['year'])>5:
        return 1000;
    
    aA = a['acousticness']
    aB = b['acousticness']
    aDistance = distance(aA, aB)
    
    bA = a['danceability']
    bB = b['danceability']
    bDistance = distance(bA, bB)
    
    cA = a['energy']
    cB = b['energy']
    cDistance = distance(cA, cB)
    
    dA = a['instrumentalness']
    dB = b['instrumentalness']
    dDistance = distance(dA, dB)
    
    eA = a['key']
    eB = b['key']
    eDistance = distance(eA, eB)
    
    fA = a['liveness']
    fB = b['liveness']
    fDistance = distance(fA, fB)
    
    gA = a['loudness']
    gB = b['loudness']
    gDistance = distance(gA, gB)
    
    hA = a['popularity']
    hB = b['popularity']
    hDistance = distance(hA, hB)
    
    iA = a['speechiness']
    iB = b['speechiness']
    iDistance = distance(iA, iB)
    
    jA = a['tempo']
    jB = b['tempo']
    jDistance = distance(jA, jB)
    
    kA = a['valence']
    kB = b['valence']
    kDistance = distance(kA, kB)
    
    return aDistance+bDistance+cDistance+dDistance+eDistance+fDistance+gDistance+hDistance+iDistance+jDistance+kDistance
    #return (-0.59*aDistance)+(0.22*bDistance)+(0.5*cDistance)+(-0.3*dDistance)+(0.011*eDistance)+(-0.075*fDistance)+(0.47*gDistance)+hDistance+(-0.14*iDistance)+(0.14*jDistance)+(0.0093*kDistance)
df = pd.read_csv('finale.csv')
df = df.drop(['Unnamed: 0','playcount','userid'],axis=1)
df = df.drop_duplicates()
query_index1=16999  #np.random.choice(df.shape[0])
print(df.iloc[query_index1]['name'])
simdict = []
for i in range(0, df.shape[0]):
    simdict.append((df.iloc[i]['name'],abs(Similarity(query_index1,i))))

```

```python
new = pd.DataFrame.from_dict(simdict)
new.columns = ["name","score"]
```
