### Importing relevant libraries

### Dataset containing the features of songs (acousticness, danceability......)

```python
spotifydf = pd.read_csv('data.csv')
```

```python
spotifydf.head()
```

```python
spotifydf = spotifydf.drop(['duration_ms','explicit','mode','release_date'],axis=1)
```

### Cleaning the spotify dataset by dropping irrelevant features above, and normalising the un-normalised features below

```python
from sklearn.preprocessing import normalize
x_array = np.array([spotifydf['tempo'],spotifydf['loudness']])
normalized_x =normalize(x_array)
spotifydf = spotifydf.drop(['tempo','loudness'],axis=1)
spotifydf.insert(3,'tempo',normalized_x[0])
spotifydf.insert(4,'loudness',normalized_x[1])
```

```python
spotifydf.head()
```

### merging the information from spotify & million song dataset based on a combo of title and artist since the song_ids are totally different

```python
combo1= np.array(spotifydf['title']+ ' '+spotifydf['artist_name'])
combo2= np.array(msddf['title']+ ' '+msddf['artist_name'])
```

```python
spotifydf=spotifydf.drop(['title','artist_name'],axis=1)
spotifydf.insert(1,'name', combo1)
msddf=msddf.drop(['title','artist_name'],axis=1)
msddf.insert(1, 'name', combo2)
```

### Generating random user id's and playcounts for entries that do not have user_info

```python
x = np.array(final['play_count'].isnull())
final['nan']= x
```

```python
final.isna().sum()
```

```python
#triplicating each entry for which the userid and play count is not available
reps = [3 if val== True else 0 for val in final.nan]
final=final.append(final.loc[np.repeat(final.index.values, reps)])
```

```python
final = final.drop(['nan'],axis=1)
```

```python
final.shape
```

```python
#generating a random playcount between 1 and 10 for each of those entries
df = pd.DataFrame(final['play_count'])
final.reset_index()
df1 = df.applymap(lambda l: l if not np.isnan(l) else np.random.choice(range(1,10)))
```

```python
df1.head(10)
```

```python
df1.shape
```

```python
final.insert(15, 'playcount', df1)
```

```python
final = final.drop(['play_count'],axis=1)
final.head(1)
```

```python
dfwithnan = pd.DataFrame(final['user_id'])
```

```python
df = pd.DataFrame(final['user_id'])
```

```python
#creating a list of unique user_ids 
df.dropna(subset = ["user_id"], inplace=True)
df.drop_duplicates(inplace=True)
```

```python
x=np.array(df['user_id'])
y= np.array(dfwithnan['user_id'])
```

```python
print(len(x)," ",len(y))
```

```python
#replacing nan values of userids with a legitimate userid, random function was taking too long and giving memory error
for i in range(0,len(y)):
    if np.isnan(y[i]):
        n=i/len(x)+1
        n = int(n)
        y[i] = x[i-n*len(x)]
```
